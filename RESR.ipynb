# loading zone distribution

```python
Zone_dist=pd.read_excel("c://Users/Dell Vostro 1014/week 39/TimeSeriesAnalysisWithPython/data_KPI/Zone (Area & Sites) distribution.xls",sheetname="Site Distribution Details")
Zone_dist.head()
```

# merging files

```python
Ini_Net_Ent=pd.merge(Ini_Net_Ent_37,pd.merge(Ini_Net_Ent_41,Ini_Net_Ent_43,how='outer'),how='outer')
rang=pd.merge(rang_37,pd.merge(rang_41,rang_43,how='outer'),how='outer')
```

# Joining the csvs

```python
resr=rang.join(Ini_Net_Ent['Number of Successful Ranging'])
resr=resr[['Start Time','Server','WBTS','Cell',"Number of Initial RNG-REQ","Number of Successful Ranging"]]
resr.head()
```

# performing Excel's 'Text to column operation'

```python
a=resr["WBTS"].str.rpartition(pat='[')
b=a[2].str.rpartition(pat=']')
resr["WBTS"]=b[0]
```

# Performing excels concat formula in pandas

```python
resr['Cell'] = resr['Cell'].astype(int)
resr['Cell'] = resr['WBTS'].str.cat(resr['Cell'].values.astype(str), sep='_')
```

```python
resr.head()
```

# Zone_dist merging

```python
resr=resr.merge(Zone_dist, on='WBTS', how='left')
```

```python
resr.head()
```

# Pivoting

```python
resr_pv=pd.pivot_table(resr,values=['Number of Initial RNG-REQ','Number of Successful Ranging'],index=['WBTS','Cell'],aggfunc=sum)
```

```python
resr_pv.head()
```

# Mathmatical operation

```python
resr_pv["RESR"]=resr_pv["Number of Successful Ranging"]*100/resr_pv["Number of Initial RNG-REQ"]
```

```python
resr_pv.head()
```

# merging

```python
resr_full=resr_pv.reset_index().merge(Zone_dist, on='WBTS', how='left')
```
